데이터를 다음과 같이 split 한다.
* data1 : 10000개의 샘플

나누는 이유는 **초기에 얻은 작은 데이터로 모델을 만든후 추가로 얻은 데이터와 비교해서 초기 모델이 얼마나 정확한지를 보기 위함이다**

```python
data = data.sample(frac=1).reset_index(drop=True) # 랜덤 셔플.
```

```python
data1 = data.loc[:10000]
data2 = data.loc[100000:]
```

```python
total1 = data1.groupby('BTH_G').count()['SEX']
normal1 = data1.groupby(['DIS','BTH_G']).count().loc[4]['SEX']
```

```python
total2 = data2.groupby('BTH_G').count()['SEX']
normal2 = data2.groupby(['DIS','BTH_G']).count().loc[4]['SEX']
```

## 2. Model

데이터를 통해서 **나이에 비례해서 얼마나 건강한지** 에 대해서 알아보려고 한다.건강에 대한 모호한 개념을 위의 데이터를 바탕으로 다음과 같이 바꾼다. <br>

$$
건강 o = 당뇨 & 혈압 진료이력이 없다
건강 x = 당뇨 or 혈압 진료이력이 있다.
$$
<br>

**건강정도**를 **나이그룹별 진료이력의 비율로** 치완해서 분석을 시작해보자!   
분석을 위해 조금더 데이터 셋을 가공해야겠다.


```python
# 나이별 정상인 비율.
health_data1= pd.DataFrame({'age': total1.index,'total':total1.ravel(),'normal':normal1.ravel()}) # data1
health_data2= pd.DataFrame({'age': total2.index,'total':total2.ravel(),'normal':normal2.ravel()}) # data2
```

가공된 데이터의 구조는 다음과같다.  
* age : 나이 그룹
* total : 해당 그룹별 인원
* noraml : 진료이력이 없는 인원

(age가 증가할수록 해당 그룹의 나이는 증가한다. 20세~)

```python
health_data1
```

```python
plt.plot(health_data1.age, health_data1.normal/health_data1.total)
print('Correlation between "age" and "normal rate" is :', np.corrcoef(health_data1.age, health_data1.normal/health_data1.total)[0][1])
```

나이가 증가할수록 건강이 나빠지고 있는것을 알수있다.  
그리고 "나이"와 "건강" 사이의 아주 강한 negatvie 상관관계를 보여주고있다.

```python
def health_data_plot(health_data1, ax=None):

    if ax is None:
        _, ax = plt.subplots()
    bg_color = ax.get_facecolor()
    rv = st.beta(health_data1.normal, health_data1.total - health_data1.normal) # rv를 고정시킴.
#     rv = st.binom(health_data1.total, health_data1.normal)
    ax.vlines(health_data1.age, *rv.interval(0.68), label=None)
    ax.plot(health_data1.age, health_data1.normal / health_data1.total, 'o', mfc=bg_color, label=None)

    ax.set_xlabel("age")
    ax.set_ylabel("healthy rate")
    ax.set_ylim(bottom=0, top=1)
    
    ax.set_xlim(left=0)
    return ax

ax = health_data_plot(health_data1);
```

로지스틱 모델을 사용해서 추론을 해보자.  
1) 건강할 비율은 나이대별로 상이하고, 2) 나이가 증가할수록 비율은 감소할 것이다

This is a very vague prior that will let the data speak for themselves.

```python
def logit_model(health_data1):
    with pm.Model() as logit_binomial:
        a = pm.Normal('a') 
        b = pm.Normal('b')

        normal = pm.Binomial('normal', 
                              n=health_data1.total, 
                              p=pm.math.invlogit(a * health_data1.age + b), 
                              observed=health_data1.normal)
    return logit_binomial


pm.model_to_graphviz(logit_model(health_data1))
```

```python
# 샘플링!
with logit_model(health_data1):
    logit_trace = pm.sample(1000, tune=1500)
pm.summary(logit_trace)
```

처음 데이터에 대한 우리의 생각과 비슷한 결과가 나왔다. 나이가 증가할수록 건강한 사람이 줄어드는 것처럼, a는 음수가 나와주었다.

```python
az.plot_posterior(logit_trace, 'a')
az.plot_posterior(logit_trace, 'b')
```

```python
ax = health_data_plot(health_data1)
t = np.linspace(0, health_data1.age.max(), 200)

for idx in np.random.randint(0, len(logit_trace), 50): # 얻은 posterior에서 각각 50개씩의 샘플을 뽑은것과 비교.
    ax.plot(t, scipy.special.expit(logit_trace['a'][idx] * t + logit_trace['b'][idx]), lw=1, color='C1', alpha=0.5)
ax.plot(t, scipy.special.expit(logit_trace['a'].mean() * t + logit_trace['b'].mean()), color='C2');
```

```python
# 커브 피팅이 잘되었는가? 얼마나 관측치에서 떨어져있는가? sample posterior predictive

with logit_model(health_data1):
    # hard to plot more than 500 sensibly
    logit_ppc = pm.sample_posterior_predictive(logit_trace, samples=500)

    
ax = health_data_plot(health_data1)
logit_ppc_normal = logit_ppc['normal'].T / health_data1.total.values.reshape(-1, 1)
t = np.linspace(0, health_data1.age.max(), 200)
ax.plot(t, scipy.special.expit(logit_trace['a'].mean() * t + logit_trace['b'].mean()), color='C2')
ax.plot(health_data1.age, logit_ppc_normal, 'k.', alpha=0.01);
```

```python
ax = pm.plot_pair(logit_trace, figsize=(7, 5));

corr = np.correlate(logit_trace['a'], logit_trace['b'])[0]
ax.set_title(f"Correlation: {corr:,.2f}");
```

```python
ax = health_data_plot(health_data1)
health_data_plot(health_data2, ax=ax)
t = np.linspace(0, health_data2.age.max(), 200)

ax.plot(t, scipy.special.expit(logit_trace['a'].mean() * t + logit_trace['b'].mean()), color='C2')
```

자 여기에 새로운 데이터로 한번 확인해보자!
기존 데이터와 새로운 데이터로 만든 모델이 얼마나 차이가 나는지 확인할 것이다.
* health_data1: 초기 데이터, 10000개 샘플
* health_data2: new 데이터, 990000개 샘플.

```python
# new data
with logit_model(health_data2):
    new_logit_trace = pm.sample(2000, tune=2000)
```

```python
ax = health_data_plot(health_data1)
health_data_plot(health_data2, ax=ax)
t = np.linspace(0, health_data2.age.max(), 200)

ax.plot(t, scipy.special.expit(logit_trace['a'].mean() * t + logit_trace['b'].mean()), color='C1', label='Trained on data')
ax.plot(t, scipy.special.expit(new_logit_trace['a'].mean() * t + new_logit_trace['b'].mean()), color='C2', label='Trained on new data')
ax.legend();
```

### 새로운 데이터로 만든 모델과 초기모델과의 차이가 크지 않다!

다시말하면 우리는 10000개의 샘플로 아주 그럴듯한 모델을 만든 것이다. 추가된 데이터들은 조금더 회귀선에 가까이 위치 해있다.  
편차가 컸던 나이 그룹의 데이터들도 잘 피팅이 되고 있음을 보여준다

실제 관측된 데이터와 posterior 모델에서 시뮬레이션된 데이터간의차이를 비교.

```python
with logit_model(health_data2):
    ppc = pm.sample_posterior_predictive(new_logit_trace)
```

```python
residuals = (health_data2.normal - ppc['normal'].mean(axis=0)) / health_data2.total

fig, ax = plt.subplots()

ax.plot(health_data2.age, residuals, 'o-')
ax.axhline(y=0, linestyle='dashed', linewidth=1)
ax.set_xlabel("age");
```
